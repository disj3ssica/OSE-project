Once a plot is generated it cannot be modified.

```python
x = [1,2,3,4]
y = [3,6,9,12]
plt.plot(x,y)
plt.show()
```

```python
# change color
plt.plot(x,y, color='purple')
plt.show
```

```python
# linewidth
plt.plot(x,y, color='purple', linewidth = 5)
plt.show
```

```python
# add marker
plt.plot(x,y, color='purple', linewidth = 1, marker='*')
plt.show
```

```python
# marker size
plt.plot(x,y, color='purple', linewidth = 1, marker='*', markersize=12)
plt.show
```

```python
# line style
plt.plot(x,y, color='purple', linewidth = 1, marker='o', markersize=12,
        linestyle='--')
plt.show
```

```python
# shorthand notation color-marker-linestyle
plt.plot(x,y,'ro--', linewidth=2, markersize=10)
plt.show()
```

```python
# add title
plt.plot(x,y,'ro--', linewidth=2, markersize=10)
plt.title('Title example')
plt.show()
```

```python
# axis labels
plt.plot(x,y,'ro--', linewidth=2, markersize=10)
plt.title('Title example')
plt.xlabel('X nickname')
plt.ylabel('Y nick')
plt.show()
```

```python
# change ticks in axes
plt.plot(x,y,'ro--', linewidth=2, markersize=10)
plt.title('Title example')
plt.xlabel('X nickname')
plt.ylabel('Y nick')

plt.xticks([1,2,3,4])
plt.yticks([3,5,7,9,11,11.5])

plt.show()
```

## Plot numpy arrays

```python
# 100 linearly spaced numbers
x = np.linspace(-5,5,100)

y = x**2
```

```python
# plot
plt.plot(x,y)
plt.show()
```

```python
# new function
y2 = x**3
```

```python
# plot multiple functions
plt.plot(x,y)
plt.plot(x,y2)
plt.show()
```

```python
# add legend
plt.plot(x,y)
plt.plot(x,y2)

plt.legend(['y=x^2','y=x^3'])

plt.show()
```

```python
plt.plot(x,y, label='y=x^2')
plt.plot(x,y2, label='y=x^3')
plt.show()
```

```python
# change axis range
plt.plot(x,y, label='y=x^2')
plt.plot(x,y2, label='y=x^3')

plt.ylim(-3,3)
plt.xlim(0,5)

plt.show()
```

```python
#change figuresize
plt.plot(x,y, label='y=x^2')
plt.plot(x,y2, label='y=x^3')

plt.ylim(-3,3)
plt.xlim(0,5)

plt.figure(figsize=(10,8), dpi=300)

plt.show()
```

```python
# SAVE
plt.plot(x,y, label='y=x^2')
plt.plot(x,y2, label='y=x^3')
plt.ylim(-3,3)
plt.xlim(0,5)
plt.figure(figsize=(10,8), dpi=300)

plt.savefig('fig example.png')

plt.show()
```

# Plotting using Pandas

```python
# load dataset
iris = pd.read_csv('https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data',
                   header=None)
columns = ['sepal_length','sepal_width', 'petal_length','petal_width','class']
iris.columns = columns
iris.head()
```

```python
# scatter plot of dataset (x,y)
iris.plot.scatter(x='sepal_length', y='sepal_width', title='Iris dataset', color='pink')
plt.show()
```

```python
# scatter plrot of data in class "Iris-setosa"
iris.groupby('class').get_group('Iris-setosa').plot.scatter(x='sepal_length', y='sepal_width',
                                                           title='Iris setosa', color='pink')
plt.show()
```

```python
# plot column values vs samples
iris.plot.line(title='Iris Dataset')
plt.show()
```

```python
# plot column values vs samples after using groupby
iris.groupby(['class']).get_group('Iris-setosa').plot.line(title='Iris setosa')
plt.show()
```

```python
# histograms of values within columns
iris['sepal_length'].plot.hist(title='Iris Dataset')
plt.show()
```

```python
# create subplots of each feature
iris.plot.hist(subplots=True, layout=(2,2), figsize=(10,10))
# 2x2 table for every feature
plt.show()
```

```python
# bar plot using groupby
iris.groupby('class')['petal_width'].mean().plot.bar()
plt.show()
```

```python
# pie charts
iris['class'].value_counts().plot.pie()
plt.show()
```

```python
# 3d surface
X, Y = np.meshgrid(np.linspace(-5,5,50), np.linspace(-5,5,50))
Z = X**2 + Y**2
```

```python
ax = plt.axes(projection='3d')
ax.plot_surface(X, Y, Z, cmap='jet')
plt.show()
```

# Contour plots

```python
# level set plot
plt.contour(X, Y, Z, cmap='jet')
plt.show()
```

```python
# fill the region
plt.contourf(X, Y, Z)
plt.show()
```
